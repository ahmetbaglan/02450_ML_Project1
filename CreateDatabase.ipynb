```python
europeFile = open('EuropeCountries.txt')
africaFile = open('AfricaCountries.txt')
asiaFile = open('AsiaCountries.txt')
nAmericaFile = open('nAmericaCountries.txt')
oceaniaFile = open('OceaniaCountries.txt')
sAmericaFile = open('sAmericaCountries.txt')
```

```python
europeCountries = europeFile.read().lower()
africaCountries = africaFile.read().lower()
asiaCountries = asiaFile.read().lower()
nAmericaCountries = nAmericaFile.read().lower()
oceaniaCountries = oceaniaFile.read().lower()
sAmericaCountries = sAmericaFile.read().lower()

for i in range(len(df)):
    c = df.iloc[i]['Country Name'].lower()
    notFound = False
    if c in europeCountries:
        df.set_value(index=i, col = 'Continent', value = 'Europe')
    elif c in africaCountries:
        df.set_value(index=i, col = 'Continent', value = 'Africa')
    elif c in asiaCountries:
        df.set_value(index=i, col = 'Continent', value = 'Asia')
    elif c in nAmericaCountries:
        df.set_value(index=i, col = 'Continent', value = 'North America')
    elif c in oceaniaCountries:
        df.set_value(index=i, col = 'Continent', value = 'Oceania')    
    elif c in sAmericaCountries:
        df.set_value(index=i, col = 'Continent', value = 'South America')
    else:
        notFound = True
    if notFound:
        print(c)
        print(i)
```
